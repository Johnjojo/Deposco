```python
import pandas as pd
import json 
```

```python
df = pd.read_csv('Louisville_Metro_KY_-_Parks_Golf_Sales_Detail.csv',low_memory=False)
df["ActualPrice"] = pd.to_numeric(df["ActualPrice"], errors='coerce').fillna(0.0).astype(int)
df["Quantity"] = pd.to_numeric(df["Quantity"], errors='coerce').fillna(0.0).astype(int)
df
```

```python
df['sale'] = df['ActualPrice'] * df['Quantity']
TOTAL_SALE_AMOUNT = (df['sale']).sum()
TOTAL_SALE_AMOUNT
```

```python
BEST_SELL_PRODUCT = df.groupby('ItemDescription')['Quantity'].sum().sort_values(ascending=False)
BEST_SELL_PRODUCT
```

```python
df['SaleDateYearMonth'] = pd.to_datetime(df['SaleDate']).apply(lambda t: t.replace(tzinfo=None)).dt.to_period('M')
MONTHLY_SALE = df.groupby(['SaleDateYearMonth'])["sale"].sum()
MONTHLY_SALE
```

```python
df
```

```python

json_result = {}
json_result['total_sale'] = int(TOTAL_SALE_AMOUNT)

new_collection = [] 

for k , v in MONTHLY_SALE.items():
    ms = {'dateYearMonth': str(k),
             'sale' : v}
    new_collection.append(ms)

json_result['monthly_sale'] = new_collection

new_collection = [] 
for k , v in BEST_SELL_PRODUCT.items():
    ps = {'product': str(k),
             'quantity' : v} 
    new_collection.append(ps)

json_result['best_sale_product'] = new_collection
```

```python
with open('data.json', 'w') as f:
    json.dump(json_result, f)
```
